# Data inladen

```python
r = requests.get('https://www.top40.nl/robots.txt')
print(r.content.decode())
```

```python
url = 'https://www.top40.nl/top40'

r = requests.get(url)

if not r.ok:
    print('Request niet gelukt')

with open('top40.html', 'wb') as f:
    f.write(r.content)
```
